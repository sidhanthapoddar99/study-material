```python
import nltk
from nltk import word_tokenize
from nltk.util import ngrams
from collections import Counter
import numpy as np
import pandas as pd
```

```python

text = "The recognition of handwritten text is challenging as there are virtually infinite ways a human can write the same message. Deep learning approaches for handwriting analysis have recently demonstrated break- through performance using both lexicon-based architectures and recurrent neural networks. This pa- per presents a fully convolutional network architecture which outputs arbitrary length symbol streams from handwritten text. A preprocessing step normalizes input blocks to a canonical representation which negates the need for costly recurrent symbol alignment correction. When a lexicon is known, we further introduce a probabilistic character error rate to correct errant word blocks. Our multi-state convolutional method is the first to demonstrate state-of-the-art results on both lexicon-based and arbitrary symbol based handwriting recognition benchmarks."
text=text.lower()

token = nltk.word_tokenize(text)
for i in range(len(token)):
    if token[i] in list(".!?;"):
        token[i]="<s>"
unigrams= ngrams(token,1)       
bigrams = ngrams(token,2)
trigrams = ngrams(token,3)
fourgrams = ngrams(token,4)
fivegrams = ngrams(token,5)
```

```python
testdata = "fully convolutional network architecture is used to outputs arbitrary symbol".split(' ')
sets=list(set(testdata))
sets
```

# Bigram martix

```python
datab=Counter(bigrams)
```

```python
arrb=[[0 for i in range (len(sets))]for j in range (len(sets))]
for i in range(len(sets)):
    for j in range(len(sets)):
        arrb[i][j]=datab[(sets[i],sets[j])]
```

```python
dfb=pd.DataFrame(np.array(arrb), index=sets,columns=sets)
dfb
```

```python
dfbp=dfb
datauni=Counter(unigrams)
for i in range(len(sets)):
    if datauni[(sets[i],)]!=0:
        dfbp.iloc[i,:]=dfbp.iloc[i,:]/datauni[(sets[i],)]
dfbp        
```

```python
dfb=pd.DataFrame(np.array(arrb), index=sets,columns=sets)
dfbp=dfb
datauni=Counter(unigrams)
for i in range(len(sets)):
    dfbp.iloc[i,:]=dfbp.iloc[i,:]+1
dfbp      
```

```python
for i in range(len(sets)):
    dfbp.iloc[i,:]=dfbp.iloc[i,:]/(datauni[(sets[i],)]+len(set(token)))
dfbp        
```
