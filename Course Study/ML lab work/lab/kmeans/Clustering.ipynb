<p>
<strong>K-Means Clustering:</strong>
K means which classifies samples based on attributes/features into K number of clusters. Clustering or grouping of samples is done by minimizing the distance between sample and the centroid
<br><br>
<strong>Guassian Mixture Clustering:</strong>
Used when uncertain about the data points where they belong or to which group. It uses probability of a sample to determine the feasibility of it belonging to a cluster.

# Applying Kmeans Clustering

### For K=4 data points

```python
#Initializing KMeans
kmeans = KMeans(n_clusters=4)
# Fitting with inputs
kmeans = kmeans.fit(data)
# Predicting the clusters
labels = kmeans.predict(data)
# Getting the cluster centers
C = kmeans.cluster_centers_
```

```python
labels
```

```python
data['labels KMeans']=labels
```

```python
ax = sns.scatterplot(x="Weight", y="Height", hue="labels KMeans", data=data)
```

```python
ax = sns.scatterplot(x="Height", y="Weight", hue="labels KMeans", data=data)
```

### For K=6

```python
#Initializing KMeans
kmeans = KMeans(n_clusters=6)
# Fitting with inputs
kmeans = kmeans.fit(data)
# Predicting the clusters
labels = kmeans.predict(data)
data['Kmeans6']=labels
```

### For K=8

```python
#Initializing KMeans
kmeans = KMeans(n_clusters=8)
# Fitting with inputs
kmeans = kmeans.fit(data)
# Predicting the clusters
labels = kmeans.predict(data)
data['Kmeans8']=labels
#ax = sns.scatterplot(x="Weight", y="Height", hue="Kmeans8", data=data)
```

```python
plt.subplot(2,1,1)
ax = sns.scatterplot(x="Weight", y="Height", hue="Kmeans6" ,data=data)

plt.subplot(2,1,2)
ax = sns.scatterplot(x="Weight", y="Height", hue="Kmeans8", data=data)
```

# Applying GMM Clustering

#### importing libraries

```python
from sklearn.mixture import GaussianMixture
```

### For K=4 data points

```python
#Initializing GMM
gmm = GaussianMixture(n_components=4)
# Fitting with inputs
gmm = gmm.fit(data)
# Predicting the clusters
labels = gmm.predict(data)
```

```python
labels
```

```python
data['labels GMM']=labels
```

```python
ax = sns.scatterplot(x="Weight", y="Height", hue="labels GMM", data=data)
```

```python
ax = sns.scatterplot(y="Weight", x="Height", hue="labels GMM", data=data)
```

### For K=6 data points

```python
#Initializing GMM
gmm = GaussianMixture(n_components=6)
# Fitting with inputs
gmm = gmm.fit(data)
# Predicting the clusters
labels = gmm.predict(data)
data['GMM6']=labels
#ax = sns.scatterplot(x="Weight", y="Height", hue="GMM6", data=data)
```

### For K=8 data points

```python
#Initializing GMM
gmm = GaussianMixture(n_components=8)
# Fitting with inputs
gmm = gmm.fit(data)
# Predicting the clusters
labels = gmm.predict(data)
data['GMM8']=labels
#ax = sns.scatterplot(x="Weight", y="Height", hue="GMM8", data=data)
```

```python
#plt.figure(figsize=(15,30))
plt.subplot(2,1,1)
ax = sns.scatterplot(x="Weight", y="Height", hue="GMM6" ,data=data)

plt.subplot(2,1,2)
ax = sns.scatterplot(x="Weight", y="Height", hue="GMM8", data=data)
```
