# RR with At

```python

def RR(bt,at,q):
    def addtoP(p1,at,bt,n,t):
        c=0
        for i in range(n,len(bt)):
            if(at[i]>t):
                return c
            else:
                p1+=[bt[i]]
                c+=1
        return c       


    noep=0
    nofp=0
    #at=[0]*5
    #bt=[15, 5, 25,12,30]
    t=0
    #q=10
    wt=0
    sbt=sum(bt)
    p1=[]
    tot=1
    while(noep<len(bt) or t<sbt):
        i=0
        noep+=addtoP(p1,at,bt,noep,tot)
        #print(noep)
        entered=False
        while i<noep:
            #print("ent")
            #print(p1)
            if(p1[i]!=0):
                #print(p1)
                #print("ent2")
                entered=True
                if(p1[i]>q):
                    for j in range(q):
                        wt+=(noep-nofp-1)
                        #print("values",noep," ",nofp)
                        p1[i]-=1
                        t+=1
                        tot+=1
                        noep+=addtoP(p1,at,bt,noep,t)
                else:
                    for j in range(p1[i]):     
                        wt+=(noep-nofp-1)
                        #print("values",noep," ",nofp)
                        p1[i]-=1
                        t+=1
                        tot+=1

                        noep+=addtoP(p1,at,bt,noep,t)
                    nofp+=1    

            i+=1

        if(not entered):
            tot+=1
    return(wt/len(bt))
```

# IRRVQ with AT

# AMRR with AT

```python
import math

def AMRR(bt,at,q):
            
    def addtoP(p1,at,bt,n,t):
        c=0
        for i in range(n,len(bt)):
            if(at[i]>t):
                return c
            else:
                p1+=[bt[i]]
                c+=1
        return c       


    noep=0
    nofp=0
    #at=[0]*5
    #bt=[15, 5, 25,12,30]
    t=0
    #q=10
    wt=0
    sbt=sum(bt)
    p1=[]
    tot=1
    noep+=addtoP(p1,at,bt,noep,tot)
    #p1.sort()
    #q=p1[nofp]
    while(noep<len(bt) or t<sbt):
    
        i=0
        noep+=addtoP(p1,at,bt,noep,tot)
        p1.sort()
        q=max(p1)+sum(p1)/(noep-nofp)
        q=int(math.ceil(q/2))
        #print(q)
        #print(q)
        #print(noep)
        entered=False
        while i<noep:
            #print("ent")
            #print(p1)
            if(p1[i]!=0):
                #print(p1) 
                #print("ent2")
                entered=True
                if(p1[i]>q):
                    for j in range(q):
                        t+=1
                        tot+=1
                        #noep+=addtoP(p1,at,bt,noep,t)
                        wt+=(noep-nofp-1)
                        #print("values",noep," ",nofp)
                        p1[i]-=1
                        noep+=addtoP(p1,at,bt,noep,t)
                    
                else:
                    for j in range(p1[i]):
                        t+=1
                        tot+=1
                        #noep+=addtoP(p1,at,bt,noep,t)
                        wt+=(noep-nofp-1)
                        #print("values",noep," ",nofp)
                        p1[i]-=1
                        noep+=addtoP(p1,at,bt,noep,t)
                    nofp+=1    

            i+=1

        if(not entered):
            tot+=1
    return(wt/len(bt))
```

# ENRICHED ROUND ROBIN ALGORITHM

```python
import math

def ERR(bt,at,q):
            
    def addtoP(p1,at,bt,n,t):
        c=0
        for i in range(n,len(bt)):
            if(at[i]>t):
                return c
            else:
                p1+=[bt[i]]
                c+=1
        return c       


    noep=0
    nofp=0
    #at=[0]*5
    #bt=[15, 5, 25,12,30]
    t=0
    #q=10
    wt=0
    sbt=sum(bt)
    p1=[]
    tot=1
    noep+=addtoP(p1,at,bt,noep,tot)
    p1.sort()
    #q=p1[nofp]
    q=sum(bt)/len(bt)
    q=q*3/4
    q=int(math.floor(q))
    #print(q)
    while(noep<len(bt) or t<sbt):
    
        i=0
        noep+=addtoP(p1,at,bt,noep,tot)
        #p1.sort()
        """q=sum(p1)/(noep-nofp)
        q=q*3/4
        q=int(math.ceil(q))"""
        
        #print(q)
        #print(p1)
        #print(noep)
        entered=False
        while i<noep:
            #print("ent")
            #print(p1)
            #print(wt)
            if(p1[i]!=0):
                #print(p1) 
                #print("ent2")
                entered=True
                if(p1[i]>=q):
                    for j in range(q):
                        t+=1
                        tot+=1
                        #noep+=addtoP(p1,at,bt,noep,t)
                        wt+=(noep-nofp-1)
                        #print("values",noep," ",nofp)
                        p1[i]-=1
                        noep+=addtoP(p1,at,bt,noep,t)
                        
                if(p1[i]==0):
                    nofp+=1
                    
                if (p1[i]<q and p1[i]!=0):
                    for j in range(p1[i]):
                        t+=1
                        tot+=1
                        #noep+=addtoP(p1,at,bt,noep,t)
                        wt+=(noep-nofp-1)
                        #print("values",noep," ",nofp)
                        p1[i]-=1
                        noep+=addtoP(p1,at,bt,noep,t)
                    nofp+=1    

            i+=1

        if(not entered):
            tot+=1
    return(wt/len(bt))
```
