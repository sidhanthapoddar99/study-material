```python
from sklearn.ensemble import RandomForestRegressor as RF
import pandas as pd
import numpy
```

```python
def predict(m,n):
    
    inp_str = bin(m)[2:]
    res =''
    for i in inp_str :
        for j in range(n):
            res+=i
    ans = int('0b'+res,2)
    return ans
```

```python
c = predict(3,2)
```

```python
l=[]
for i in range(1,100000):
    if i not in range(2000,3500):
        l=l+[[i,predict(i,2)]]
```
